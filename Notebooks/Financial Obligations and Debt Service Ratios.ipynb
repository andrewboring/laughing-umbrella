```python
import pandas as pd
```

```python
#file = "../Data/Debt_Service/FRB_FOR.csv"
url = "https://storage.api.eriscape.com/v1/AUTH_andrew/laughing-umbrella/Data/Debt_Service/FRB_FOR.csv"
```

```python
data = pd.read_csv(url)
```

```python
data.head(10)
```

```python
df = data.drop([0, 1, 2, 3, 4])
```

```python
df.head()
```

```python
df.rename(columns={'Series Description':'Period'}, inplace=True)
```

```python
df.head()
```

### Resample monthly and interpolate values

```python
import datetime
df['Period'] = pd.to_datetime(df['Period'])
df.set_index('Period',inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
# If we don't convert to numeric dtypes, the Interpolate below won't work.
# We couldn't do this when reading CSV, as the columns had mixed data toward the top.

df = df.astype({'Financial obligations ratio, seasonally adjusted':'float64', 
                'Consumer debt service ratio, seasonally adjusted':'float64',
                'Debt service ratio, seasonally adjusted':'float64', 
                'Mortgage debt service ratio, seasonally adjusted':'float64'}) 
```

## Upsample and Interpolate.

```python
resampled_df = df.resample('MS').first()
```

```python
resampled_df.dtypes
```

```python
#resampled_df.drop(columns=['Quarter'])
#new_df
```

```python
resampled_df = resampled_df.interpolate(method='linear')
```

```python
#upsampled = df.resample('M')
#interpolated = upsampled.interpolate(method='linear')
#interpolated
```

```python
yet_another_df = resampled_df.reset_index()
#yet_another_df = yet_another_df.astype({'Financial obligations ratio, seasonally adjusted':'float64',)
yet_another_df.head()
```

```python
yet_another_df.info()
```

```python
yet_another_df['Year'] = yet_another_df['Period'].dt.year
yet_another_df['Month'] = yet_another_df['Period'].dt.month.astype

# Extract Year from Period
#yet_another_df['Year'] = yet_another_df['Period'].str.extract('([0-9]+)',expand=False).str.strip()

# Extract Quarter from Period
#df['Quarter'] = df['Period'].str.extract('(Q\d)',expand=False).str.strip()
#yet_another_df['Month'] = yet_another_df['Month'].apply(str)
```

```python
#yet_another_df['Month'] = yet_another_df['Period'].dt.month
yet_another_df.head()
```

```python
final_df = yet_another_df.drop(columns=['Period'])
final_df.head()
```

```python
final_df['Month'] = final_df['Month'].map({'1':'January', 
                                '2':'February',
                                '3':'March',
                                '4':'April',
                                '5':'May',
                                '6':'June',
                                '7':'July',
                                '8':'August',
                                '9':'September',
                                '10':'October',
                                '11':'November',
                                '12':'December',
                                })
```
