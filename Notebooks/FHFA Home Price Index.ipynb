
# FHFA Home Price Index


Source: Federal Housing Finance Agency House Price Index (HPI), Purchase-only, Monthly Historical  

Links:
 - https://www.fhfa.gov/DataTools/Downloads/Pages/House-Price-Index.aspx
 - https://www.fhfa.gov/Media/PublicAffairs/Pages/House-Price-Index-Frequently-Asked-Questions.aspx
 

```python
import pandas as pd
```

```python
#file = "../Data/FHFA/HPI_PO_monthly_hist.xls"
url = "https://storage.api.eriscape.com/v1/AUTH_andrew/laughing-umbrella/Data/FHFA/HPI_PO_monthly_hist.xls"
data = pd.read_excel(url, header=2)
```

```python
data.head(5)
```

```python
#Remove "non-seasonally adjusted" columns, identified by "(NSA)" in the name.
df = data[data.columns.drop(list(data.filter(regex='NSA')))]
```

```python
df = df.drop(0)
```

```python
# Extract Year/Month from Period
df.rename(columns={'Month':'Period'}, inplace=True)
df['Year'] = df['Period'].dt.year
df['Month'] = df['Period'].dt.month
```

```python
df.head()
```

```python
df = df.drop(columns=['Period'])
```

```python
df['Month'] = df['Month'].apply(str)
```

```python
df['Month'] = df['Month'].map({'1':'January', 
                                '2':'February',
                                '3':'March',
                                '4':'April',
                                '5':'May',
                                '6':'June',
                                '7':'July',
                                '8':'August',
                                '9':'September',
                                '10':'October',
                                '11':'November',
                                '12':'December',
                                })
```
