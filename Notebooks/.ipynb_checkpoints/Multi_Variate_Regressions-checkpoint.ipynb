```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../Data/dataset.csv")
df.head()
```

```python
# Assign the data to X and y

X = df.drop(columns=["period","CS_Index"])
y = df["CS_Index"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Use train_test_split to create training and testing data

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

### END SOLUTION
```

```python
#scaler = StandardScaler().fit(X_train)
```

```python
#X_train_scaled = scaler.transform(X_train)
#X_test_scaled = scaler.transform(X_test)
```

```python
# Create the model using LinearRegression

### BEGIN SOLUTION - needed to run ElasticNet due to model over fit
from sklearn.linear_model import ElasticNet
model = ElasticNet()
### END SOLUTION
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
model.coef_
```

```python
# Plot the Residuals for the Training and Testing data



plt.scatter(model.predict(X_train), model.predict(X_train).ravel() - y_train.ravel(), c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test).ravel() - y_test.ravel(), c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
```
