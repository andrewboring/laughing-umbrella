```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
housing_data_master_load = "../Data/dataset.csv"
housing_data_master_df = pd.read_csv(housing_data_master_load)
```

```python
housing_data_master_df.head()
```

```python
housing_data_master_df.columns
```

```python
x_axis = housing_data_master_df["Housing Affordability Index"]
y_axis = housing_data_master_df["Market Absorption Rate (%)"]
size_value = housing_data_master_df["USA"]
trace0 = plt.scatter(
     x= x_axis,
     y= y_axis,
     marker= "o",
     s= size_value/10,
     label= "USA", 
     c= "orange",edgecolor= "black" 
        
     )


data = [trace0]

x_axis = housing_data_master_df["Housing Affordability Index"]
y_axis = housing_data_master_df["Market Absorption Rate (%)"]
size_value = housing_data_master_df["Pacific"]
trace1 = plt.scatter(
     x= x_axis,
     y= y_axis,
     marker= "o",
     s= size_value/10,
     label= "Pacific", 
     c= "yellow",edgecolor= "black"
        
     )


data = [trace1]

x_axis = housing_data_master_df["Housing Affordability Index"]
y_axis = housing_data_master_df["Market Absorption Rate (%)"]
size_value = housing_data_master_df["South Atlantic"]
trace2 = plt.scatter(
     x= x_axis,
     y= y_axis,
     marker= "o",
     s= size_value/10,
     label= "South Atlantic", 
     c= "lightskyblue", edgecolor= "black"
    )
data = [trace2]
#py.iplot(data, filename='bubblechart-size')
plt.xlabel ("Housing Affordability")
plt.ylabel("Market Absorption")
plt.legend(title= "USA and Region")
plt.grid(True)
plt.title("Housing Affordability Impact on Housing Sales")
#plt.text(42,35,"Note: Circle size correlates with Region")
plt.show()
```
