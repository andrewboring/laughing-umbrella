```python
import pandas as pd
```

```python
file = "../Data/New_Home_Sales/SeriesReport-201905101525.csv"
data = pd.read_csv(file, header=7)
```

```python
data.head(5)
```

```python
#Remove "non-seasonally adjusted" columns, identified by "(NSA)" in the name.
df = data[['Period','Value']]
```

```python
df.head()
```

```python
# Extract Year/Month from Period
df['Year'] = df['Period'].str.extract('([0-9]+)',expand=False).str.strip()
df['Month'] = df['Period'].str.extract('([A-Za-z]+)',expand=False).str.strip()
```

```python
df.head()
```

```python
df = df.drop(columns=['Period'])
```

```python
df['Month'] = df['Month'].apply(str)
```

```python
df['Month'] = df['Month'].map({'Jan':'January', 
                                'Feb':'February',
                                'Mar':'March',
                                'Apr':'April',
                                'May':'May',
                                'Jun':'June',
                                'Jul':'July',
                                'Aug':'August',
                                'Sep':'September',
                                'Oct':'October',
                                'Nov':'November',
                                'Dec':'December',
                                })
```

```python
df
```

```python
final_df = df.rename(columns={'Value':'Market Absorption Rate (%)'})
```
